```python
def equal(a,b,c):
    numbers = [a,b,c]
    res = numbers.count(max(numbers, key=numbers.count)) # returns most repeated value
    return res if res > 1 else 0 # if most repeated value is repeated more than once return it
```

```python
equal(1,2,2)
```
